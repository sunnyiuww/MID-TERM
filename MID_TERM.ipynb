```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

**Penjelasan Kode**

1. **`cols = df.select_dtypes(include=['float64', 'int64'])`**: Memilih hanya kolom numerik bertipe `float64` dan `int64` untuk diproses.
  
2. **Looping setiap kolom**:

   - **`iqr` (Interquartile Range)**: Menghitung rentang antar-kuartil sebagai `q3 - q1`.
   - **`lower_bound` dan `upper_bound`**: Menentukan batas bawah (`q1 - 1.5*iqr`) dan batas atas (`q3 + 1.5*iqr`) untuk mendeteksi outliers.
   - **Outliers**: Menyaring baris-baris yang memiliki nilai di bawah `lower_bound` atau di atas `upper_bound`.
   - **Persentase Outliers**: Menghitung persentase outliers dengan membagi jumlah outliers dengan total baris `df`, lalu mengalikan dengan 100.

3. **`results.append(...)`**: Menambahkan nama kolom dan persentase outliers ke dalam list `results`.

Inconsistent Values

```python
# Fungsi untuk memeriksa nilai yang tidak konsisten di semua kolom
def check_inconsistent_values_by_dtype(df):
    inconsistent_values_dict = {}

    for column in df.columns:
        # Mendapatkan tipe data kolom
        expected_dtype = df[column].dtype

        if expected_dtype == 'int64':
            # Memeriksa nilai yang tidak dapat dikonversi menjadi integer
            inconsistent_values = df[column][df[column].apply(lambda x: not isinstance(x, int))]

        elif expected_dtype == 'float64':
            # Memeriksa nilai yang tidak dapat dikonversi menjadi float
            inconsistent_values = df[column][df[column].apply(lambda x: not isinstance(x, (int, float)))]

        else:
            # Kolom dengan tipe data lain (misalnya, objek/string)
            continue

        if not inconsistent_values.empty:
            inconsistent_values_dict[column] = inconsistent_values.tolist()

    return inconsistent_values_dict

# Memeriksa nilai yang tidak konsisten di semua kolom
inconsistent_values_found = check_inconsistent_values_by_dtype(df)

# Menampilkan hasil
if inconsistent_values_found:
    for column, values in inconsistent_values_found.items():
        print(f"Nilai yang tidak konsisten dalam kolom '{column}': {values}")
else:
    print("Tidak ada nilai yang tidak konsisten ditemukan di semua kolom.")
```

**Penjelasan Kode:**

1. **Fungsi `check_inconsistent_values_by_dtype`**:
   - Membuat dictionary `inconsistent_values_dict` untuk menyimpan nilai-nilai yang tidak konsisten yang ditemukan di setiap kolom.
   - Untuk setiap kolom dalam `df`, memeriksa tipe data yang diharapkan (dengan `expected_dtype`):
     - Jika tipe datanya `int64`, fungsi mencari nilai yang bukan integer menggunakan `apply(lambda x: not isinstance(x, int))`.
     - Jika tipe datanya `float64`, fungsi mencari nilai yang bukan integer atau float menggunakan `apply(lambda x: not isinstance(x, (int, float)))`.
   - Untuk tipe data lain (seperti string atau objek), pengecekan dilewati.
   - Jika ada nilai yang tidak konsisten (tidak sesuai tipe data kolom), nilai tersebut akan ditambahkan ke dictionary `inconsistent_values_dict` dalam bentuk list.

2. **Memeriksa Nilai Tidak Konsisten**:
   - Fungsi `check_inconsistent_values_by_dtype(df)` dijalankan dan hasilnya disimpan di `inconsistent_values_found`.
   
3. **Menampilkan Hasil**:
   - Jika `inconsistent_values_found` tidak kosong, kode akan mencetak nilai yang tidak konsisten untuk setiap kolom yang memiliki nilai tersebut.
   - Jika tidak ada nilai yang tidak konsisten, maka mencetak pesan bahwa semua nilai sesuai tipe data.

#####**Imputasi Outliers**

```python
# Membuat list untuk menyimpan hasil perhitungan
results = []

# Memilih kolom dengan tipe data numerik
cols = df.select_dtypes(include=['float64', 'int64'])

# Perhitungan outliers
for col in cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Imputasi outliers menggunakan metode flooring dan capping
    df[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    percent_outliers = (len(outliers) / len(df)) * 100
    results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})
```

**Penjelasan Kode:**

**Deteksi dan Imputasi Outliers**

1. **Inisialisasi Hasil**:
   - Sebuah list `results` dibuat untuk menyimpan hasil perhitungan persentase outliers untuk setiap kolom.

2. **Memilih Kolom Numerik**:
   - Kolom-kolom yang memiliki tipe data numerik (baik `float64` maupun `int64`) dipilih dengan menggunakan fungsi `select_dtypes`.

3. **Loop untuk Mendeteksi Outlier**:
   - Untuk setiap kolom dalam list kolom numerik (`cols`), langkah-langkah berikut dilakukan:
     - **Menghitung Kuartil**:
       - Kuartil pertama (`q1`) dan kuartil ketiga (`q3`) dihitung untuk mendapatkan rentang interkuartil (IQR).
     - **Menghitung IQR**:
       - IQR dihitung sebagai selisih antara `q3` dan `q1`.
     - **Menentukan Batasan**:
       - `lower_bound` dan `upper_bound` ditentukan menggunakan rumus yang umum digunakan untuk mendeteksi outliers:
         - `lower_bound` adalah `q1 - 1.5 * IQR`
         - `upper_bound` adalah `q3 + 1.5 * IQR`
     - **Imputasi Outliers**:
       - Outliers ditangani dengan metode *flooring* dan *capping*:
         - Setiap nilai yang lebih kecil dari `lower_bound` diubah menjadi nilai `lower_bound`.
         - Setiap nilai yang lebih besar dari `upper_bound` diubah menjadi nilai `upper_bound`.
       - Ini dilakukan dengan menggunakan fungsi `apply` dan lambda function untuk setiap elemen dalam kolom.
     - **Mendapatkan Outliers Setelah Imputasi**:
       - Setelah imputasi, baris yang masih dianggap outlier (jika ada) dikumpulkan, dan persentase dari jumlah outliers tersebut dihitung.
     - **Menyimpan Hasil**:
       - Hasil perhitungan (nama kolom dan persentase outliers) disimpan dalam list `results`.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

**Penjelasan Kode:**

Kode yang digunakan sama dengan yang digunakan pada tahap pengecekkan outliers, digunakan kembali hanya untuk melakukan pengecekkan setelah tahap imputasi outliers.

###**Penanganan Inconsistent Values**

#####**Pengecekkan Inconsistent Values**

```python
# Fungsi untuk memeriksa nilai yang tidak konsisten di semua kolom
def check_inconsistent_values_by_dtype(df):
    inconsistent_values_dict = {}

    for column in df.columns:
        # Mendapatkan tipe data kolom
        expected_dtype = df[column].dtype

        if expected_dtype == 'int64':
            # Memeriksa nilai yang tidak dapat dikonversi menjadi integer
            inconsistent_values = df[column][df[column].apply(lambda x: not isinstance(x, int))]

        elif expected_dtype == 'float64':
            # Memeriksa nilai yang tidak dapat dikonversi menjadi float
            inconsistent_values = df[column][df[column].apply(lambda x: not isinstance(x, (int, float)))]

        else:
            # Kolom dengan tipe data lain (misalnya, objek/string)
            continue

        if not inconsistent_values.empty:
            inconsistent_values_dict[column] = inconsistent_values.tolist()

    return inconsistent_values_dict

# Memeriksa nilai yang tidak konsisten di semua kolom
inconsistent_values_found = check_inconsistent_values_by_dtype(df)

# Menampilkan hasil
if inconsistent_values_found:
    for column, values in inconsistent_values_found.items():
        print(f"Nilai yang tidak konsisten dalam kolom '{column}': {values}")
else:
    print("Tidak ada nilai yang tidak konsisten ditemukan di semua kolom.")
```

**Penjelasan Kode:**

**Fungsi Memeriksa Nilai Tidak Konsisten**

1. **Inisialisasi Kamus Hasil**:
   - Sebuah dictionary bernama `inconsistent_values_dict` dibuat untuk menyimpan hasil pemeriksaan nilai yang tidak konsisten untuk setiap kolom.

2. **Loop Melalui Kolom**:
   - Fungsi ini iterasi melalui setiap kolom di DataFrame.
   - Untuk setiap kolom, tipe data yang diharapkan (`expected_dtype`) diperoleh.

3. **Memeriksa Tipe Data**:
   - **Untuk Kolom Tipe `int64`**:
     - Menggunakan `apply` untuk memeriksa apakah setiap elemen dalam kolom adalah tipe integer. Jika ada nilai yang tidak dapat dikonversi menjadi integer, nilai-nilai tersebut akan disimpan.
   - **Untuk Kolom Tipe `float64`**:
     - Demikian juga, kolom tipe float diperiksa untuk memastikan bahwa semua nilai adalah tipe integer atau float. Nilai yang tidak memenuhi syarat disimpan.
   - **Untuk Tipe Data Lain**:
     - Kolom dengan tipe data lain, seperti objek atau string, diabaikan dalam pemeriksaan ini.

4. **Menyimpan Hasil**:
   - Jika ada nilai yang tidak konsisten ditemukan dalam kolom, nilai-nilai tersebut disimpan dalam dictionary `inconsistent_values_dict`, di mana kunci adalah nama kolom dan nilainya adalah daftar nilai yang tidak konsisten.

5. **Mengembalikan Hasil**:
   - Fungsi mengembalikan dictionary `inconsistent_values_dict` yang berisi informasi tentang kolom-kolom dengan nilai yang tidak konsisten.

**Menampilkan Hasil**

- Setelah memanggil fungsi, jika ada nilai yang tidak konsisten ditemukan, hasilnya ditampilkan dengan mencetak nama kolom dan nilai yang tidak konsisten.
- Jika tidak ada nilai yang tidak konsisten ditemukan, pesan yang menyatakan hal tersebut akan dicetak.

**Penjelasan Output:**

Karena tidak ada inconsistent values yang diidentifikasi, maka tidak perlu ada penanganan yang dilakukan.

#A/B TESTING

```python
# Hipotesis
H0 = "Rata-rata ADR dari City Hotel sama dengan Resort Hotel."
H1 = "Rata-rata ADR dari City Hotel berbeda dengan Resort Hotel."

# Memisahkan ADR berdasarkan jenis hotel
city_hotel_adr = df[df['hotel'] == 'City Hotel']['adr']
resort_hotel_adr = df[df['hotel'] == 'Resort Hotel']['adr']

# Melakukan uji Independent T-test
t_stat, p_value = stats.ttest_ind(city_hotel_adr, resort_hotel_adr)

# Menyusun hasil uji dalam DataFrame
ab_testing_results = pd.DataFrame({
    'T-statistic': [t_stat],
    'P-value': [p_value],
    'Significant Difference': [p_value < 0.05]  # Jika p-value < 0.05, kita menolak H0
})

# Membuat visualisasi Box Plot dengan Plotly
fig = px.box(df, x='hotel', y='adr', color='hotel',
             title='Distribusi ADR Berdasarkan Jenis Hotel',
             labels={'adr': 'ADR (Average Daily Rate)', 'hotel': 'Jenis Hotel'})

# Menambahkan garis mean pada box plot
fig.update_traces(boxmean='sd')  # Menampilkan rata-rata dan deviasi standar

# Menampilkan visualisasi
fig.show()

# Menampilkan hasil uji t-test
print("Hipotesis")
print(f"H0: {H0}")
print(f"H1: {H1}\n")

print("### Hasil Uji A/B Testing (T-test) ###")
display(ab_testing_results)

# Keputusan berdasarkan P-value
if p_value < 0.05:
    print("\nKeputusan: Menolak H0. Karena terdapat perbedaan signifikan dalam rata-rata ADR antara City Hotel dan Resort Hotel.")
else:
    print("\nKeputusan: Gagal menolak H0. Karena tidak terdapat perbedaan signifikan dalam rata-rata ADR antara City Hotel dan Resort Hotel.")
```

**Penjelasan Kode:**

**Hipotesis**
- **H0** (Hipotesis Nol): Rata-rata ADR dari City Hotel sama dengan Resort Hotel.
- **H1** (Hipotesis Alternatif): Rata-rata ADR dari City Hotel berbeda dengan Resort Hotel.

**1. Memisahkan Data Berdasarkan Jenis Hotel**
- Data ADR dipisahkan berdasarkan jenis hotel:
  - `city_hotel_adr`: Mengambil nilai ADR untuk City Hotel.
  - `resort_hotel_adr`: Mengambil nilai ADR untuk Resort Hotel.

**2. Melakukan Uji Independent T-test**
- Uji T dilakukan untuk membandingkan rata-rata antara dua kelompok (City Hotel dan Resort Hotel):
  - `stats.ttest_ind(city_hotel_adr, resort_hotel_adr)`: Menghitung statistik T dan nilai p untuk uji T independen.
  - `t_stat`: Nilai statistik T yang menunjukkan seberapa besar perbedaan antara rata-rata kedua kelompok.
  - `p_value`: Nilai p yang menunjukkan probabilitas untuk mendapatkan hasil yang sama atau lebih ekstrem jika H0 benar.

**3. Menyusun Hasil Uji dalam DataFrame**
- Hasil uji T disusun dalam DataFrame:
  - Terdapat kolom untuk T-statistic, P-value, dan informasi apakah perbedaan signifikan berdasarkan P-value.

**4. Visualisasi Distribusi ADR**
- Menggunakan Plotly untuk membuat box plot yang menggambarkan distribusi ADR berdasarkan jenis hotel:
  - `px.box()`: Membuat box plot dengan sumbu X adalah jenis hotel dan sumbu Y adalah ADR.
  - `fig.update_traces(boxmean='sd')`: Menambahkan garis rata-rata dan deviasi standar pada box plot untuk memberikan informasi tambahan.

**5. Menampilkan Hasil Uji T-test**
- Menampilkan hipotesis dan hasil uji T-test dengan menggunakan `print` dan `display()`.

**6. Keputusan Berdasarkan P-value**
- Jika nilai p kurang dari 0.05:
  - Menolak H0, yang berarti ada perbedaan signifikan dalam rata-rata ADR antara City Hotel dan Resort Hotel.
- Jika nilai p lebih besar dari atau sama dengan 0.05:
  - Gagal menolak H0, yang berarti tidak ada perbedaan signifikan dalam rata-rata ADR antara kedua jenis hotel.

Uji A/B testing atau **T-test** digunakan untuk membandingkan rata-rata dari dua kelompok (dalam hal ini, **City Hotel** dan **Resort Hotel**) untuk menentukan apakah ada perbedaan yang signifikan secara statistik antara keduanya. Berikut adalah penjelasan dari hasil uji A/B tersebut:
